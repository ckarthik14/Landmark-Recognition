```python
fd = open("landmark-ids.txt","r")
L = []
for line in fd:
    line = line.strip()
    L.append(line)
```

```python
import pandas as pd
df = pd.read_csv("google-landmarks-dataset/train.csv")
df2 = pd.read_csv("google-landmarks-dataset/boxes_split2.csv")
```

```python
merged_df = pd.merge(df, df2, on=['id'], how='inner')
merged_df
```

```python
final_df = merged_df[merged_df['landmark_id'].isin(L)]
final_df = final_df.drop(columns=['url', 'landmark_id'])
```

```python
final_df.to_csv("modified_box_split_2.csv",index=False)
```
